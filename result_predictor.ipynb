```python
total = match.merge(ball,on='ID')
```

```python
total = total[total['innings'] == 2]
```

```python
total.columns
```

```python
total.shape
```

```python
total = total.drop(columns =['Unnamed: 0','Team1Players','Team2Players','Umpire1', 'Umpire2','Venue'] )
```

```python
total
```

```python
total = total.rename(columns = {"total_run_y":"runs","total_run_x":"total_run"})
```

```python
total.head()
```

```python
total["cur_score"]=total.groupby(["ID","innings"]).cumsum()["runs"]
```

```python
total
```

```python
total = total[total['innings'] == 2]
```

```python
total
```

```python
total["runs_needed"]=total["total_run"]-total["cur_score"]
```

```python
total
```

```python
total["overs"]=total["overs"]+1
```

```python
total['balls_left'] = 126 - (total['overs']*6 + total['ballnumber'])
```

```python
total
```

```python
total.columns
```

```python
total["isWicketDelivery"]
```

```python
total["wicket_fallen"] = total.groupby("ID").cumsum()["isWicketDelivery"]
```

```python
total
```

```python
total['wick_in_hand'] = 10 - total["wicket_fallen"]
```

```python
total
```

### current run rate

```python
total["crr"]=(total["cur_score"]*6/(120 - total["balls_left"])).round(2)
```

### required run rate

```python
total["rr"]=total["runs_needed"]*6/total["balls_left"]
```

```python
total
```

### generating output columns

```python
def result(ip):
    if ip["BattingTeam"]==ip["WinningTeam"]:
        return 1
    else:
        return 0
```

```python
total["outcome"]=total.apply(result,axis=1)
```

```python
total.head(2)
```

```python
def team2(dff,f1,f2,f3):
    team2 = []
    for i in range(dff.shape[0]):
        if dff[f1].iloc[i]==dff[f2].iloc[i]:
            team2.append(dff[f3].iloc[i])
        else:
            team2.append(dff[f2].iloc[i])
    return team2
    
```

```python
total["BowlingTeam"]=team2(total,"BattingTeam","Team1","Team2")
```

```python
total.columns
```

```python
final = total[['BattingTeam','BowlingTeam','City','runs_needed','balls_left','wick_in_hand','total_run','crr','rr','outcome']]
```

```python
final = final[final['balls_left'] != 0]
```

```python
final
```

```python
final= final.sample(final.shape[0])
```

```python
final
```

```python
final.dropna(inplace=True)
```

```python
x = final.iloc[:,:-1]
y = final.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=16)
```

```python
x_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler,OneHotEncoder
```

```python
ct = ColumnTransformer([("tr1",OneHotEncoder(sparse_output=False,drop = "first"),['BattingTeam','BowlingTeam','City']),
                       ],remainder = "passthrough")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline(steps = [("st1",ct),
                        ("st2",LogisticRegression())])
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred = pipe.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
